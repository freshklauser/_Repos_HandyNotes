```python
# .isalnum(): 如果 string 至少有一个字符并且所有字符都是字母或数字则返回 True,否则返回 False
substr = 'hello89today001world'
substr.isalnum()
```

```python
# .isalpha(): 如果 string 至少有一个字符并且所有字符都是字母则返回 True,否则返回 False
'asbd'.isalpha()
```

```python
# .istitle(): 如果 string 是标题化的(见 title())则返回 True，否则返回 False
# title(): 首字母大写，其余全是小写
s = ['aBC', 'Bcd', 'BdN']
print([ele.istitle() for ele in s])
S = [ele.title() for ele in s]
print(S)
```

```python
# .isspace(): 如果 string 中只包含空格（包含换行和tab等），则返回 True，否则返回 False.
s = ['  ', 's  ', '\t  ', '\n']
print([ele.isspace() for ele in s])
```

```python
'>>>>>>>>>>>>>>>> python中str函数isdigit、isdecimal、isnumeric的区别 <<<<<<<<<<<<<<'
# .isdigit(): 如果 string 只包含数字则返回 True 否则返回 False.
# .isdecimal(): 如果 string 只包含 `十进制数字` 则返回 True 否则返回 False.
# .isnumeric(): 如果 string 中只包含数字字符，则返回 True，否则返回 False
```

```python
num = "1"  # unicode or 全角
print(num.isdigit(), num.isdecimal(), num.isnumeric())
```

```python
num = "四" # 汉字
print(num.isdigit(), num.isdecimal(), num.isnumeric())
```

```python
num = "Ⅷ" # 罗马数字
print(num.isdigit(), num.isdecimal(), num.isnumeric())
```

```python
num = "1.235" # 小数
print(num.isdigit(), num.isdecimal(), num.isnumeric())
```

```python
# 非10进制
print('0x1e94f'.isdecimal(), '0b101'.isdigit(), '0b101'.isnumeric())    # 16进制数字返回False
```

```python
num = b"1" # byte
try:
    print('.isdigit(): ', num.isdigit())
    # True, Error, Error
except Exception as e:
    print(e)
try:
    print('.isdecimal(): ', num.isdecimal())
    # True, Error, Error 
except Exception as e:
    print(e)
try:
    print('.isnumeric(): ', num.isnumeric())
    # True, Error, Error
except Exception as e:
    print(e)
```

### 2) 字符串的字母大小写系列

```python
# .islower(): 全是小写返回True
# .isupper(): 全是大写返回True
s = 'aaa'
print(s.islower(), s.isupper())
```

```python
# .lower():      转换 string 中所有大写字符为小写.
# .upper():      转换 string 中的小写字母为大写
# .title():      所有单词都是以大写开始，其余字母均为小写(见 istitle())
# .capitalize(): 把字符串的第一个字符大写 ( == s.title())
# .swapcese():   翻转 string 中的大小写
s = 'abCDeF'
print(s.lower(), s.upper(), s.title(), s.capitalize(), s.swapcase())
```

### 3） 字符串对齐和填充至固定长度系列

```python
# .center(width):  返回一个原字符串居中,并使用空格填充至长度 width 的新字符串
# .ljust(width):   返回一个原字符串左对齐,并使用空格填充至长度 width 的新字符串
# .rjust(width):
# .zfill(width):   返回长度为 width 的字符串，原字符串 string 右对齐，前面填充0
s = 'handypoints'
width = 20
print('空格填充--中间对齐：', '>>>', s.center(width), '<<<')
print('空格填充--左边对齐：', '>>>', s.ljust(width), '<<<')
print('空格填充--右边对齐：', '>>>', s.rjust(width), '<<<')
print('零值填充--右边对齐：', '>>>', s.zfill(width), '<<<')
```
